### Explanation of the export code

`summons_sample` is the dataframe we want to export

`.to_csv()` saves our dataframe to a .csv file. You are baseically exporting your python data out into a format that Excel, Google Sheets can read.

`index=False`
By default, on export pandas includes the DataFrame's index – the row numbers (0, 1, 2, 3...). We NEVER
need it 99.99 percent of the time. `index=False` simply tells pandas to leave that column out on export.

`encoding="utf-8"`
Tells Python which character encoding to use when writing the file. `UTF-8` is the most universal standard — it handles English plus accented characters, symbols, non-Latin scripts, etc. without garbling them. If your data has names like "García" or "Müller" and you skip this, you might get weird characters or errors depending on your system.

### Prompt: 

Provide the pandas code to calculate which boroughs had the most summons (in descending order).


````python
# 1. Which borough had the most summons? 

# Replace string '(null)' placeholders with real NaN so they're excluded cleanly
nypd_summons['BORO'] = nypd_summons['BORO'].replace('(null)', pd.NA)

# Count summons per borough, sort descending, and drop any missing borough values
borough_counts = (
    nypd_summons['BORO']
    .value_counts()              # Count occurrences of each borough; sorts descending by default
    .reset_index()               # Convert the Series into a two-column DataFrame
    .rename(columns={            # Rename columns from pandas defaults to readable names
        'BORO': 'borough',
        'count': 'summons_count'
    })
)

borough_counts
````

````python
# 2. What were the top 10 most common offenses?

# prompt in gemini : "Provide the pandas code to calculate what were the top 10 most common offenses (in descending order)."

# Calculate the top 10 most common offenses
# value_counts() counts occurrences and sorts them in descending order by default
# head(10) retrieves the first 10 rows from the sorted result
top_10_offenses = nypd_summons['OFFENSE_DESCRIPTION'].value_counts().head(10)

# Get the top 10 offenses

# Display the result
print("Top 10 most common offenses (Descending Order):")

print(top_10_offenses) # Output the top 10 offenses to the console
````

````python
# 4. What were some of the most unusual summons?

## prompt in gemini: "Provide the pandas code to determine the offenses with only one summons"

# Calculate the frequency of each offense
offense_counts = nypd_summons['OFFENSE_DESCRIPTION'].value_counts() # Count occurrences of each offense

# Filter for offenses that occur exactly once
# This identifies unique or single-occurrence summons types
single_summons_offenses = offense_counts[offense_counts == 1] # Keep only offenses with count equal to 1

# Display the result
print("Most Unusual summons:")
print(single_summons_offenses) # Output the list of unique offenses to the console
````

````python
#7. Which borough had the fewest offenses per 100K people?

## prompt in gemini: "Provide the pandas code to determine Which borough had the fewest offenses per 100K people? (in descending order)

# 2020 Census population
populations = {
    "BROOKLYN": 2736074,
    "QUEENS": 2405464,
    "MANHATTAN": 1694251,
    "BRONX": 1472654,
    "STATEN ISLAND": 495747
}

# Calculate the count of summons per borough
borough_counts = nypd_summons['BORO'].value_counts() # Count occurrences of each borough

# Convert the counts into a DataFrame for easier calculation
offense_stats = borough_counts.to_frame(name='offense_count') # Create DataFrame from series

# Filter the DataFrame to include only the boroughs present in the populations dictionary
offense_stats = offense_stats[offense_stats.index.isin(populations.keys())] # Keep only relevant boroughs

# Map the population data to the corresponding boroughs
offense_stats['population'] = offense_stats.index.map(populations) # Add population column

# Calculate the number of offenses per 100,000 people
# Formula: (Number of Offenses / Population) * 100,000
offense_stats['offenses_per_100k'] = (offense_stats['offense_count'] / offense_stats['population']) * 100000 # Calculate rate

# Sort the results in descending order of the offenses per 100k
offense_stats_sorted = offense_stats.sort_values(by='offenses_per_100k', ascending=False) # Sort by calculated rate

# Display the final results
print("Offenses per 100K people by borough (Descending Order):")
print(offense_stats_sorted[['offense_count', 'offenses_per_100k']]) # Output the resulting table
````
